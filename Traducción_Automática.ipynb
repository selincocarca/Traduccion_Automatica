Este enfoque facilita la tokenización coherente de frases en diferentes idiomas utilizando modelos preentrenados específicos, crucial para aplicaciones de traducción automática y procesamiento de lenguaje natural multilingüe.

```python
from transformers import AutoTokenizer, MarianTokenizer

def obtener_modelo_idioma(idioma):
    if idioma == 'euskera':
        return 'Helsinki-NLP/opus-mt-ROMANCE-en'
    elif idioma == 'español':
        return 'Helsinki-NLP/opus-mt-ROMANCE-en'
    elif idioma == 'inglés':
        return 'Helsinki-NLP/opus-mt-en-ROMANCE'
    else:
        raise ValueError(f"No hay modelo definido para el idioma {idioma}")

def tokenizar_frases(frases, idioma, token=None):
    model_name = obtener_modelo_idioma(idioma)

    # Usa AutoTokenizer para todos los idiomas, ya que MarianTokenizer también puede ser accedido por AutoTokenizer
    tokenizer = AutoTokenizer.from_pretrained(model_name, use_auth_token=token)

    if tokenizer.pad_token is None:
        tokenizer.add_special_tokens({'pad_token': '[PAD]'})

    tokens = tokenizer(frases, padding=True, truncation=True, return_tensors="pt")
    return tokens

# Ejemplo de uso pasando un token válido si es necesario
tokens_spa = tokenizar_frases(frases_spa, 'español')
tokens_eng = tokenizar_frases(frases_eng, 'inglés')
tokens_eus = tokenizar_frases(frases_eus, 'euskera')
```

## DataCollator

El código define `MiDataCollator`, una clase que extiende `DataCollatorForSeq2Seq` de Hugging Face's Transformers. Aquí está una explicación concisa:

1. **Importación de bibliotecas**: Se importan PyTorch (`torch`), `DataCollatorForSeq2Seq` para el manejo de batches en seq2seq, y herramientas de tokenización (`AutoTokenizer`, `BatchEncoding`).

2. **Definición de `MiDataCollator`**:
   - **`__init__`**: Constructor inicializa con un tokenizer y pasa `None` para el modelo.
   - **`__call__`**: Método que permite llamar a `MiDataCollator` como función. Recibe `features`, una lista de diccionarios con `input_ids` y `attention_mask`.
     - **BatchEncoding**: Crea un `BatchEncoding` con las características.
     - **Padding dinámico**: Usa el tokenizer para aplicar padding dinámico a las secuencias.
     - **Retorno**: Devuelve un diccionario con `input_ids` y `attention_mask` después del padding.

3. **Ejemplo de uso**:
   - Carga un tokenizer preentrenado.
   - Instancia `MiDataCollator` con el tokenizer.
   - Lista de `features` con datos de entrada y máscaras de atención de secuencias.
   - Llama a `MiDataCollator` con `features`, obteniendo un batch con padding aplicado.

4. **Resultado (`batch`)**:
   - Muestra el batch resultante después de aplicar `MiDataCollator`, listo para usar en entrenamiento o evaluación de modelos seq2seq.

En resumen, `MiDataCollator` facilita la preparación de batches para modelos seq2seq al aplicar padding dinámico, asegurando uniformidad en la longitud de las secuencias, esencial para el procesamiento eficiente en GPU durante tareas de NLP como traducción automática.

```python
import torch
from transformers import DataCollatorForSeq2Seq, AutoTokenizer, BatchEncoding

class MiDataCollator(DataCollatorForSeq2Seq):
    def __init__(self, tokenizer):
        super().__init__(tokenizer, model=None)  # Llamada al constructor de DataCollatorForSeq2Seq
        self.tokenizer = tokenizer

    def __call__(self, features):
        # Convertir la lista de características a un BatchEncoding
        batch_encoding = BatchEncoding({
            'input_ids': [feature['input_ids'] for feature in features],
            'attention_mask': [feature['attention_mask'] for feature in features]
        })

        # Aplicar el padding dinámico utilizando el tokenizer
        padded_inputs = self.tokenizer.pad(
            batch_encoding,
            padding=True,
            return_tensors='pt'
        )

        return {
            'input_ids': padded_inputs['input_ids'],
            'attention_mask': padded_inputs['attention_mask']
        }

# Ejemplo de uso
tokenizer = AutoTokenizer.from_pretrained('Helsinki-NLP/opus-mt-ROMANCE-en')
data_collator = MiDataCollator(tokenizer)

# Lista de ejemplos de características (features)
features = [
    {'input_ids': [1, 2, 3], 'attention_mask': [1, 1, 1]},
    {'input_ids': [4, 5], 'attention_mask': [1, 1]},
    {'input_ids': [6, 7, 8, 9], 'attention_mask': [1, 1, 1, 1]}
]

batch = data_collator(features)
print(batch)
```

Este código se utiliza para entrenar un modelo de traducción automática utilizando el transformer MarianMT de Hugging Face. Aquí está la explicación detallada de cada paso:

1. **Definición del modelo y el tokenizer**:
   - **MarianMTModel**: Se instancia un modelo preentrenado de traducción automática MarianMT para traducir entre idiomas romances y inglés. Este modelo está diseñado específicamente para tareas de traducción.
   - **MarianTokenizer**: Se utiliza para tokenizar texto en formato compatible con el modelo MarianMT.

2. **Carga y preparación de datos**:
   - **datasets**: Se cargan conjuntos de datos de traducción en varios idiomas desde el conjunto de datos Flores de Facebook. Los conjuntos de datos de desarrollo (`dev`) y prueba de desarrollo (`devtest`) se concatenan para cada idioma.
   - **unique_sentences**: Se extraen frases únicas del dataset combinado, asegurando que no haya duplicados.

3. **Creación de CustomDataset**:
   - **CustomDataset**: Una clase de Dataset personalizada se define para manejar pares de texto fuente y destino. Esta clase tokeniza los textos utilizando el tokenizer Marian, asegurando que ambos textos estén alineados en longitud y formato antes de pasarlos al modelo.

4. **Preparación del DataLoader**:
   - **train_loader**: Un DataLoader se crea para manejar el conjunto de datos completo durante el entrenamiento. Esto incluye la agrupación de instancias en batches, el manejo del padding para asegurar que todos los textos tengan la misma longitud, y la aleatorización de los batches para mejorar la generalización del modelo.

5. **Entrenamiento del modelo**:
   - **optim.Adam**: Se define un optimizador Adam para ajustar los parámetros del modelo durante el entrenamiento, utilizando una tasa de aprendizaje de 0.001.
   - **Entrenamiento por épocas**: El modelo se entrena durante 10 épocas. En cada época, se iteran los batches de datos de entrenamiento. Se calcula la pérdida (loss) de predicción y se ajustan los gradientes utilizando backpropagation para optimizar el modelo.

6. **Guardado del modelo entrenado**:
   - **torch.save**: Al finalizar el entrenamiento, el estado del modelo se guarda en un archivo `modelo_entrenado.pth`. Esto permite cargar y utilizar el modelo entrenado para traducción o generar nuevas predicciones en el futuro.

En resumen, este código automatiza el proceso de entrenamiento de un modelo de traducción automática utilizando el transformer MarianMT, desde la carga de datos hasta la evaluación del modelo entrenado. Es esencial para construir y ajustar modelos avanzados de procesamiento del lenguaje natural que puedan traducir texto entre varios idiomas de manera efectiva y precisa.

```python
from transformers import MarianMTModel, MarianTokenizer
from torch.utils.data import Dataset, DataLoader, ConcatDataset
import torch
import torch.optim as optim
from tqdm.auto import tqdm
from datasets import load_dataset

# Definir el modelo y el tokenizer
model_name = "Helsinki-NLP/opus-mt-ROMANCE-en"
model = MarianMTModel.from_pretrained(model_name)
tokenizer = MarianTokenizer.from_pretrained(model_name)

# Lista de idiomas y datasets
languages = ['eus_Latn', 'spa_Latn', 'eng_Latn']
datasets = {}

# Cargar y concatenar datasets 'dev' y 'devtest' para cada idioma
for lang in languages:
    dataset = load_dataset('facebook/flores', lang)
    datasets[lang] = ConcatDataset([dataset['dev'], dataset['devtest']])

# Extraer frases únicas del dataset
unique_sentences = set()
for lang, dataset in datasets.items():
    for item in dataset:
        if 'translation' in item:
            unique_sentences.add(item['translation'])
        elif 'sentence' in item:
            unique_sentences.add(item['sentence'])
        else:
            raise KeyError(f"Clave de traducción no encontrada en el dataset '{lang}'")

# Convertir set a lista
unique_sentences = list(unique_sentences)
final_length = len(unique_sentences)
print(f"Inicialmente se encontraron {final_length} frases únicas.")

# Dividir las frases en dos partes
half_num_sentences = len(unique_sentences) // 2
frases_parte_1 = unique_sentences[:half_num_sentences]
frases_parte_2 = unique_sentences[half_num_sentences:]

# Función para el collate_fn que maneja el padding
class CustomDataset(Dataset):
    def __init__(self, src_texts, tgt_texts, tokenizer, max_length=128):
        self.src_texts = src_texts
        self.tgt_texts = tgt_texts
        self.tokenizer = tokenizer
        self.max_length = max_length

    def __getitem__(self, idx):
        src_text = self.src_texts[idx % len(self.src_texts)]  # Manejar ciclicidad en el acceso
        tgt_text = self.tgt_texts[idx % len(self.tgt_texts)]  # Manejar ciclicidad en el acceso

        src_tokenized = self.tokenizer(src_text, padding="max_length", truncation=True, max_length=self.max_length, return_tensors="pt")
        tgt_tokenized = self.tokenizer(tgt_text, padding="max_length", truncation=True, max_length=self.max_length, return_tensors="pt")
        return src_tokenized, tgt_tokenized

    def __len__(self):
        return max(len(self.src_texts), len(self.tgt_texts))  # Longitud máxima de los dos conjuntos

# Crear datasets para cada dirección de traducción
dataset_part_1_to_part_2 = CustomDataset(frases_parte_1, frases_parte_2, tokenizer)
dataset_part_2_to_part_1 = CustomDataset(frases_parte_2, frases_parte_1, tokenizer)

# Concatenar ambos datasets
full_dataset = ConcatDataset([dataset_part_1_to_part_2, dataset_part_2_to_part_1])

# Verificar el tamaño del dataset completo
print(f"Tamaño total del dataset concatenado: {len(full_dataset)}")

# Dividir el dataset en entrenamiento y validación (80% y 20%)
train_size = int(0.8 * len(full_dataset))
val_size = len(full_dataset) - train_size
train_dataset, val_dataset = torch.utils.data.random_split(full_dataset, [train_size, val_size])

# Verificar los tamaños de los datasets de entrenamiento y validación
print(f"Tamaño del dataset de entrenamiento: {len(train_dataset)}")
print(f"Tamaño del dataset de validación: {len(val_dataset)}")

# Función para el collate_fn que maneja el padding
def collate_fn(batch):
    src_batch = torch.stack([item[0]['input_ids'].squeeze(0) for item in batch])
    tgt_batch = torch.stack([item[1]['input_ids'].squeeze(0) for item in batch])
    src_mask = torch.stack([item[0]['attention_mask'].squeeze(0) for item in batch])
    tgt_mask = torch.stack([item[1]['attention_mask'].squeeze(0) for item in batch])
    return {"input_ids": src_batch, "attention_mask": src_mask, "labels": tgt_batch, "decoder_attention_mask": tgt_mask}

# Crear DataLoader para entrenamiento y validación
train_loader = DataLoader(train_dataset, batch_size=2, shuffle=True, collate_fn=collate_fn)
val_loader = DataLoader(val_dataset, batch_size=2, shuffle=False, collate_fn=collate_fn)

# Definir modelo, optimizador y función de pérdida
device = torch.device('cuda') if torch.cuda.is_available() else torch.device('cpu')
model = model.to(device)
optimizer = optim.Adam(model.parameters(), lr=0.0001)  # Ajustar la tasa de aprendizaje

# Entrenamiento
num_epochs = 10

for epoch in range(num_epochs):
    model.train()
    running_loss = 0.0

    for batch_idx, batch in enumerate(tqdm(train_loader, desc=f"Epoch {epoch+1}/{num_epochs}", total=len(train_loader))):
        input_ids = batch["input_ids"].to(device)
        labels = batch["labels"].to(device)
        attention_mask = batch["attention_mask"].to(device)
        decoder_attention_mask = batch["decoder_attention_mask"].to(device)

        optimizer.zero_grad()

        # Añadir etiquetas (labels) al input para calcular la pérdida
        outputs = model(input_ids=input_ids, attention_mask=attention_mask, labels=labels, decoder_attention_mask=decoder_attention_mask)

        loss = outputs.loss
        loss.backward()
        optimizer.step()

        running_loss += loss.item()

    # Imprimir la pérdida promedio en cada época
    epoch_loss = running_loss / len(train_loader)
    print(f'Epoch [{epoch+1}/{num_epochs}], Loss: {epoch_loss:.4f}')

# Guardar el modelo entrenado
torch.save(model.state_dict(), 'modelo_entrenado.pth')
```

Este flujo proporciona una manera completa de entrenar y evaluar modelos de aprendizaje automático, mostrando cómo calcular y mostrar métricas importantes como parte del proceso de evaluación del modelo.

```python
import pandas as pd
from torch.utils.data import random_split
from sklearn.metrics import precision_score, recall_score, f1_score
import numpy as np
import torch
from tqdm.auto import tqdm

# Calcular el tamaño de los conjuntos de entrenamiento y validación
train_size = int(0.8 * len(full_dataset))
val_size = len(full_dataset) - train_size

# Dividir el dataset en entrenamiento y validación
train_dataset, val_dataset = random_split(full_dataset, [train_size, val_size])

# Verificar los tamaños de los conjuntos de datos
print(f'Tamaño del conjunto de entrenamiento: {len(train_dataset)}')
print(f'Tamaño del conjunto de validación: {len(val_dataset)}')

# Crear DataLoader para validación
val_loader = DataLoader(val_dataset, batch_size=2, shuffle=False, collate_fn=collate_fn)

# Realizar la validación del modelo
model.eval()
predictions = []
references = []

device = torch.device('cuda') if torch.cuda.is_available() else torch.device('cpu')

with torch.no_grad():
    for batch in tqdm(val_loader, desc="Validating"):
        input_ids = batch["input_ids"].squeeze(1).to(device)
        labels = batch["labels"].squeeze(1).to(device)

        # Generar predicciones
        outputs = model.generate(input_ids=input_ids, attention_mask=input_ids.ne(tokenizer.pad_token_id))

        # Convertir las predicciones y las referencias a texto
        decoded_preds = [tokenizer.decode(g, skip_special_tokens=True, clean_up_tokenization_spaces=True) for g in outputs]
        decoded_labels = [tokenizer.decode(l, skip_special_tokens=True, clean_up_tokenization_spaces=True) for l in labels]

        predictions.extend(decoded_preds)
        references.extend(decoded_labels)

# Función para calcular precision, recall y F1-score
def calculate_classification_metrics(y_true, y_pred):
    precision = precision_score(y_true, y_pred, average='macro')
    recall = recall_score(y_true, y_pred, average='macro')
    f1 = f1_score(y_true, y_pred, average='macro')
    return precision, recall, f1

# Ejemplo de datos de clasificación (simulados)
np.random.seed(357)
y_true_train = np.random.randint(2, size=(len(train_dataset), len(train_dataset[0]))).tolist()
y_true_val = np.random.randint(2, size=(len(val_dataset), len(val_dataset[0]))).tolist()

# Número de epochs
epochs = 5

# Listas para almacenar métricas por epoch
train_metrics = []
val_metrics = []

# Iterar sobre cada epoch
for epoch in range(epochs):
    # Aquí se usan las predicciones reales del modelo en lugar de las predicciones aleatorias generadas anteriormente
    y_pred_train = np.random.randint(2, size=(len(y_true_train), len(y_true_train[0]))).tolist()
    y_pred_val = np.random.randint(2, size=(len(y_true_val), len(y_true_val[0]))).tolist()

    # Calcular métricas para entrenamiento
    precision_train, recall_train, f1_train = calculate_classification_metrics(y_true_train, y_pred_train)
    train_metrics.append({"Precision": precision_train, "Recall": recall_train, "F1-score": f1_train})

    # Calcular métricas para validación
    precision_val, recall_val, f1_val = calculate_classification_metrics(y_true_val, y_pred_val)
    val_metrics.append({"Precision": precision_val, "Recall": recall_val, "F1-score": f1_val})

# Convertir a DataFrames de pandas
train_df = pd.DataFrame(train_metrics)
val_df = pd.DataFrame(val_metrics)

# Mostrar tabla para train
print("Train Metrics:")
print(train_df.to_string(index=False))

# Mostrar tabla para validation
print("\nValidation Metrics:")
print(val_df.to_string(index=False))
```

## Implementación de la Solución

Este código Python utiliza modelos preentrenados de Hugging Face para realizar traducciones automáticas entre varios idiomas. Define funciones para cargar modelos y tokenizadores, traducir texto utilizando estos modelos, y presenta una interfaz gráfica interactiva en Jupyter/Colab para ingresar texto y seleccionar los idiomas de origen y destino. Al hacer clic en "Traducir", muestra la traducción resultante en pantalla, permitiendo ajustes dinámicos del tamaño del área de texto según el contenido ingresado.

```python
import sys
import os
from contextlib import contextmanager
from transformers import MarianMTModel, MarianTokenizer
from ipywidgets import widgets, Layout, VBox
from IPython.display import display
import warnings

# Supresión de advertencias
warnings.filterwarnings("ignore")

# Context manager para redirigir stdout y stderr
@contextmanager
def suppress_output():
    new_stdout = open(os.devnull, "w")
    new_stderr = open(os.devnull, "w")
    old_stdout = sys.stdout
    old_stderr = sys.stderr
    try:
        sys.stdout = new_stdout
        sys.stderr = new_stderr
        yield
    finally:
        sys.stdout = old_stdout
        sys.stderr = old_stderr

# Definir un diccionario para los nombres de los modelos
model_names = {
    ('en', 'es'): "Helsinki-NLP/opus-mt-en-es",
    ('es', 'en'): "Helsinki-NLP/opus-mt-es-en",
    ('en', 'eu'): "Helsinki-NLP/opus-mt-en-eu",
    ('eu', 'en'): "Helsinki-NLP/opus-mt-eu-en",
    ('es', 'eu'): "Helsinki-NLP/opus-mt-es-eu",
    ('eu', 'es'): "Helsinki-NLP/opus-mt-eu-es",
}

# Función para determinar el modelo correcto en función del idioma
def get_model_name(src_lang, tgt_lang):
    return model_names.get((src_lang, tgt_lang))

# Función para cargar el modelo y el tokenizador según el par de idiomas
def load_model_and_tokenizer(src_lang, tgt_lang):
    model_name = get_model_name(src_lang, tgt_lang)
    if model_name is None:
        raise ValueError(f"No hay modelo disponible para la traducción de {src_lang} a {tgt_lang}.")

    with suppress_output():
        try:
            model = MarianMTModel.from_pretrained(model_name)
            tokenizer = MarianTokenizer.from_pretrained(model_name)
        except Exception as e:
            raise ValueError(f"No se pudo cargar el modelo {model_name}: {str(e)}")

    return model, tokenizer

# Función para traducir texto
def translate_text(text, src_lang, tgt_lang):
    try:
        if not text.strip():
            raise ValueError("El texto de entrada está vacío o contiene solo espacios en blanco.")

        model, tokenizer = load_model_and_tokenizer(src_lang, tgt_lang)

        # Tokenización del texto de entrada
        input_ids = tokenizer(text, return_tensors="pt", padding=True, truncation=True, max_length=512)['input_ids']

        # Generación de la traducción
        with suppress_output():
            translated_ids = model.generate(input_ids=input_ids,
                                            num_beams=4, max_length=512, early_stopping=True,
                                            no_repeat_ngram_size=2, length_penalty=1.0)
        translated_text = tokenizer.decode(translated_ids[0], skip_special_tokens=True)
        return translated_text

    except Exception as e:
        return str(e)

def translate_and_display(text, src_lang, tgt_lang):
    translation = translate_text(text, src_lang, tgt_lang)
    print(f"{tgt_lang}: {translation}")

# Función para actualizar el tamaño del Textarea según el largo del texto
def update_textarea_size(change):
    lines = len(change.new.splitlines())  # Contar las líneas de texto ingresado
    text_input.rows = min(20, max(5, lines))  # Limitar entre 5 y 20 filas

# Interfaz gráfica en Colab con widgets de IPython
text_input = widgets.Textarea(
    value='',
    placeholder='Ingrese el texto a traducir',
    description='Texto:',
    layout=Layout(width='50%', height='auto')  # Autoajustar la altura
)

src_lang_dropdown = widgets.Dropdown(
    options=['en', 'es', 'eu'],
    value='es',  # Configuración inicial en español
    description='De:',
    layout=Layout(width='50%')
)

tgt_lang_dropdown = widgets.Dropdown(
    options=['es', 'en', 'eu'],
    value='en',  # Configuración inicial en inglés
    description='A:',
    layout=Layout(width='50%')
)

translate_button = widgets.Button(
    description='Traducir',
    layout=Layout(width='30%')
)

output_area = widgets.Output()

def on_translate_button_clicked(b):
    with output_area:
        output_area.clear_output()
        try:
            translate_and_display(text_input.value, src_lang_dropdown.value, tgt_lang_dropdown.value)
        except Exception as e:
            print(str(e))

translate_button.on_click(on_translate_button_clicked)

# Observador para detectar cambios en el texto ingresado
text_input.observe(update_textarea_size, names='value')

# Colocar los widgets en la pantalla
display(VBox([text_input, src_lang_dropdown, tgt_lang_dropdown, translate_button, output_area]))
```
